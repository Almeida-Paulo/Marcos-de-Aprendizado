```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('datasets/kc_house_data.csv')
```

```python
dataset.drop(['id', 'date'], axis = 1, inplace=True)
```

```python
dataset['renovated'] = dataset['yr_renovated'].apply(lambda x : 1 if x!=0 else 0)
dataset.drop('yr_renovated', axis=1, inplace=True)
```

```python
sns.set(style = 'whitegrid',
        font = 'Rubik', 
        font_scale = 1.1, 
        rc={'axes.facecolor':'#2F4858', 'figure.facecolor':'w', "axes.spines.right": False, "axes.spines.top": False})
```

```python
plt.figure(figsize=(5, 7))
sns.displot(x='price', data=dataset, bins=100, kde=True, height=7, aspect=1.5, color='#4BBC8E')
plt.ticklabel_format(style='plain', axis='x')
plt.title("Price Distribution", fontsize=16)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize = (13, 10))

sns.boxplot(ax=axes[0], x='bedrooms', y='price', data=dataset, width=0.9, palette='BuGn_r')
axes[0].ticklabel_format(style='plain', axis='y')
axes[0].set_xlabel('Bedrooms', fontsize=13)
axes[0].set_ylabel('Price', fontsize=13)
axes[0].set_title('Bedrooms vs Price', fontsize=16)

sns.countplot(ax=axes[1], x='bedrooms', data=dataset, palette='BuGn_r')
axes[1].set_xlabel('Bedrooms', fontsize=13)
axes[1].set_title("Bedroom's count", fontsize=16)
plt.subplots_adjust(wspace=0.4, hspace=0.4)
```
